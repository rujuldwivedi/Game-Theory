```python
import nashpy as nash
import numpy as np

# Alice's payoff matrix
A = np.array([
    [2, 0],   # Alice chooses Opera or Football
    [0, 1]
])

# Bob's payoff matrix
B = np.array([
    [1, 0],
    [0, 2]
])

# Create the game
bos = nash.Game(A, B)
print("ðŸŽ­ Battle of the Sexes Game Created!\n")

# Show payoff matrices
print("Alice's Payoffs:\n", A)
print("Bob's Payoffs:\n", B)

# Enumerate all Nash equilibria
print("\nðŸŽ¯ Nash Equilibria (Support Enumeration):")
for eq in bos.support_enumeration():
    alice, bob = eq
    print(f"Alice: Opera={alice[0]:.2f}, Football={alice[1]:.2f}")
    print(f"Bob:   Opera={bob[0]:.2f}, Football={bob[1]:.2f}")
    print("-" * 40)
```

```python
import random

# Mixed strategy probabilities
alice_probs = {"Opera": 2/3, "Football": 1/3}
bob_probs   = {"Opera": 1/3, "Football": 2/3}
```

```python
def choose_action(prob_dist):
    r = random.random()
    cumulative = 0
    for action, prob in prob_dist.items():
        cumulative += prob
        if r <= cumulative:
            return action
```

```python
def simulate_battle(rounds=50):
    coordination = 0
    misalignment = 0
    alice_opera = 0
    bob_football = 0

    print(f"\nðŸŽ­ Simulating {rounds} Rounds of Battle of the Sexes:\n")

    for i in range(1, rounds + 1):
        alice_choice = choose_action(alice_probs)
        bob_choice   = choose_action(bob_probs)

        if alice_choice == "Opera":
            alice_opera += 1
        if bob_choice == "Football":
            bob_football += 1

        aligned = (alice_choice == bob_choice)
        if aligned:
            coordination += 1
            result = "âœ… Coordinated"
        else:
            misalignment += 1
            result = "âŒ Misaligned"

        print(f"Round {i}: Alice â†’ {alice_choice}, Bob â†’ {bob_choice} â†’ {result}")

    print("\nðŸ“Š Final Summary:")
    print(f"Coordinated: {coordination} times")
    print(f"Misaligned:  {misalignment} times")
    print(f"Alice chose Opera:   {alice_opera} times")
    print(f"Bob chose Football:  {bob_football} times")

    print("\nðŸŽ¯ Coordination Rate: {:.2f}%".format((coordination / rounds) * 100))

if __name__ == "__main__":
    simulate_battle(30)
```

```python
import random
import matplotlib.pyplot as plt

# Mixed strategy probabilities
alice_probs = {"Opera": 2/3, "Football": 1/3}
bob_probs   = {"Opera": 1/3, "Football": 2/3}
```

```python
def choose_action(prob_dist):
    r = random.random()
    cumulative = 0
    for action, prob in prob_dist.items():
        cumulative += prob
        if r <= cumulative:
            return action
```

```python
def simulate_battle_with_visuals(rounds=50):
    coordination = 0
    misalignment = 0
    alice_opera = 0
    bob_football = 0

    alice_choices = []
    bob_choices = []
    coordination_trend = []

    print(f"\nðŸŽ­ Simulating {rounds} Rounds of Battle of the Sexes:\n")

    for i in range(1, rounds + 1):
        alice_choice = choose_action(alice_probs)
        bob_choice   = choose_action(bob_probs)

        alice_choices.append(alice_choice)
        bob_choices.append(bob_choice)

        aligned = (alice_choice == bob_choice)
        if aligned:
            coordination += 1
            result = "âœ… Coordinated"
        else:
            misalignment += 1
            result = "âŒ Misaligned"

        coordination_trend.append(coordination)

        if alice_choice == "Opera":
            alice_opera += 1
        if bob_choice == "Football":
            bob_football += 1

        print(f"Round {i}: Alice â†’ {alice_choice}, Bob â†’ {bob_choice} â†’ {result}")

    # Summary
    print("\nðŸ“Š Final Summary:")
    print(f"Coordinated: {coordination} times")
    print(f"Misaligned:  {misalignment} times")
    print(f"Alice chose Opera:   {alice_opera} times")
    print(f"Bob chose Football:  {bob_football} times")
    print("\nðŸŽ¯ Coordination Rate: {:.2f}%".format((coordination / rounds) * 100))

    # === VISUALIZATIONS ===
    # Line plot: Coordination trend
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, rounds + 1), coordination_trend, label="Cumulative Coordination", marker='o', color='green')
    plt.title("ðŸ“ˆ Coordination Trend Over Time")
    plt.xlabel("Round Number")
    plt.ylabel("Total Coordinated Rounds")
    plt.grid(True)
    plt.tight_layout()
    plt.legend()
    plt.show()

    # Pie Charts: Player choices
    def show_pie(data, title):
        counts = {"Opera": data.count("Opera"), "Football": data.count("Football")}
        labels = counts.keys()
        values = counts.values()
        plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=90, colors=["#8ecae6", "#ffb703"])
        plt.axis("equal")
        plt.title(title)
        plt.show()

    show_pie(alice_choices, "ðŸŽ­ Alice's Choices")
    show_pie(bob_choices, "ðŸŸï¸ Bob's Choices")

# Run simulation with visuals
if __name__ == "__main__":
    simulate_battle_with_visuals(30)
```
